```python
import spotipy
import spotipy.util as util
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
client_id= "" #Aqui, o usuário deve informar o id da conta spotify
client_secret= "" # Inserir token de segurança
redirect_uri=""
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id,
                                                           client_secret=client_secret))
```

```python
artist_id = ""
```

```python
discos = sp.artist_albums(artist_id, album_type="album", country="BR", limit=11, offset=0)
```

```python
albuns = discos['items']
```

```python
table = pd.DataFrame(data=albuns)
table
```

```python
table.dtypes
```

```python
table.drop(columns=['album_group','external_urls','href','type'], inplace=True)
```

```python
table
```

```python
discos = table['id']
discos
```

```python
albums_ids = table['id'].to_list()
albums_ids
```

```python
print(type(albums_ids), type(discos))
```

```python
[print(i) for i in albums_ids]
```

```python
for i in range(len(albums_ids)):
    tracks = (sp.album_tracks(albums_ids[i], market="BR"))
```

```python
tracks
```

```python
musicas = tracks['items']
```

```python
musicas = pd.DataFrame(data=musicas)
musicas[['sp','s1','id_track']] = (musicas['uri'].str.split(':', expand = True))
```

```python
musicas.dtypes
```

```python
musicas
```

```python
musicas['id_track']
```

```python
analise_tracks = sp.audio_features(tracks=musicas['id_track'])
analise_tracks = pd.DataFrame(data=analise_tracks)
analise_tracks
```

```python
#Aqui vamos juntar os títulos de cada música com seu id respectivo
analise = pd.merge(analise_tracks, musicas, on = 'id', how = 'outer')
analise
```

```python
ax = sns.scatterplot(data = analise, x = "valence", y = "name", hue = "valence", palette="blend:#003f5c,#f95d6a")
ax.set(xlabel='tristeza - positividade',
       ylabel='Música',
       title='Análise de sentimento')
```

```python
sns.color_palette("light:#5A9", as_cmap=True)
```
